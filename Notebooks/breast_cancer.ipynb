```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Diagnosis'),
                                               df['Diagnosis'],
                                               test_size=0.2,
                                               random_state=42,
                                               stratify=df['Diagnosis'])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
ctf=ColumnTransformer([('ss',StandardScaler(),num_col)],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

```python
models={
    'LogisticRegression':LogisticRegression(max_iter=5000,class_weight='balanced',random_state=42,solver='saga'),
    'DecisionTree':DecisionTreeClassifier(random_state=42),
    'RandomForest':RandomForestClassifier(n_estimators=50,n_jobs=-1,random_state=42),
    'XGBoost':XGBClassifier(random_state=42),
    'SVM':SVC(kernel='rbf',C=1.0,gamma='scale',probability=True,random_state=42)
}
```

```python
for name,model in models.items():
    pipe=Pipeline([
        ('preprocessing',ctf),
        ('model',model)
    ])
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    acc=accuracy_score(y_pred,y_test)
    cv=cross_val_score(pipe,df.drop(columns=['Diagnosis']),df['Diagnosis'],cv=5,scoring='accuracy')
    print(f'{name}')
    print(f'accuracy:{acc:.4f}')
    print(f'CV average:{cv.mean():.4f}')
    print(classification_report(y_test,y_pred))
```

Logistic Regression is giving best results

```python
pipe_lr=Pipeline([
    ('preprocessing',ctf),
    ('lr',LogisticRegression(max_iter=5000,class_weight='balanced',random_state=42,solver='saga'))
])
```

```python
pipe_lr.fit(x_train,y_train)
```

```python
y_pred=pipe_lr.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```
